# INDICO.IO sentiment_hq test
Passing data as a batch and separately returns different results.
Using the NUS SMS corpus as test data.

Can attribute some of the erroneous value to Singlish.

```python
import indicoio
import json
from tabulate import tabulate

indicoio.config.api_key = open('indico_api.txt', 'r').read()

f = open('smsCorpus_en_2015.03.09_all.json', 'r').read()
data = json.loads(f)
batch = []
individual = []
# just getting the first 20 texts.

for i in range(100):
    val = data['smsCorpus']['message'][i]['text']['$']
    batch.append(val)
    print('*', end='')
    individual.append(indicoio.sentiment_hq(val))
print()

batch_res = indicoio.sentiment_hq(batch)
both = zip(batch_res, individual)

print(tabulate(both, headers=['Batch', 'Individual']))
```

```python
%matplotlib inline
import matplotlib
from matplotlib import pyplot as plt

matplotlib.rcParams['figure.figsize'] = [12.0, 8.0]

plt.plot(batch_res)
plt.plot(individual)
plt.show()
```
